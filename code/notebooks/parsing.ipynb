# Parsing filtered_df.csv

```python
import pandas as pd
import numpy as np
import pathlib
import sys, os

print(os.path.abspath(''))
sys.path.insert(0, str(pathlib.Path(os.path.abspath('')).resolve().parents[0]))
import paths
```

```python
paths.show_paths()  # Verificar que las rutas se han configurado correctamente
df = pd.read_csv(paths.DATA_PATH/"df_filtered.csv")
df
```

```python
df = df[['subject','trial','session', 'date', 'x_c', 'r_c', 'ttype_n', 'stimd_n', 'performance', 'timepoint_1', 'timepoint_2', 'timepoint_3', 'timepoint_4', 'onset', 'offset']]
```

```python
df['stim_d'] = df['offset'] - df['onset']
df['delay_d'] = df['timepoint_4'] - df['offset']

response_code = {'L': 0, 'C': 1, 'R': 2}

df['response'] = df['r_c'].map(response_code)
df['stimulus'] = df['x_c'].map(response_code)

df.sort_values(['subject', 'session', 'trial', 'date'], inplace=True)
df['trial_idx'] = df.groupby(['subject']).cumcount()
```

```python
df.to_parquet(paths.DATA_PATH/"df_filtered.parquet", index=False)
```
